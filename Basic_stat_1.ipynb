```python
 #quantile 1
 print('quantile 1:',df['Volume'].quantile(0.25))
 #quantile 2
 print('quantile 1:',df['Volume'].quantile(0.50))
  #quantile 3
print('quantile 1:',df['Volume'].quantile(0.75))
```

```python
#Inter Quantile range

print("IQR",df['Volume'].quantile(0.75)-df['Volume'].quantile(0.25))

#Upper Limit
print('Upper Limit:',6+1.5*(6-3))
print('Lower Limit:',6-1.5*(6-3))
```

```python
Num_Columns= ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Amount', 'Net Sales Value']
for i in Num_Columns:
  Column_name = i
  print("Outlier Caluculations for",i)
  Q1=df[Column_name].quantile(0.25)
  print('Quartile 1: ', Q1)
  Q2=df[Column_name].quantile(0.50)
  print('Quartile 2/ Median: ', Q2)
  Q3=df[Column_name].quantile(0.75)
  print('Quartile 3: ', Q3)
  #Interquartile Range
  IQR=Q3-Q1
  #Upper Limit
  UL=Q3+(1.5*IQR)
  print('Upper Limit: ', UL)
  #Lower Limt
  LL = Q1-(1.5*IQR)
  print('Lower Limit: ', LL)
  Outliers=df[(df[Column_name]>UL) | (df[Column_name]<LL)]
  print(Outliers)

  print('________________')
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
print(categorical_columns)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Calculate SKU frequencies
Brand_counts = df['Brand'].value_counts()
threshold = 5
df['Brand_grouped'] =df['Brand'].apply(lambda x: x if Brand_counts[x] >= threshold else 'Other')
sns.countplot(x='Brand_grouped', data=df, order=df['Brand_grouped'].value_counts().index)
plt.title('Frequency of Brand (Grouped)')
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
BU_counts = df['BU'].value_counts()
threshold = 5
df['BU_grouped'] =df['BU'].apply(lambda x: x if BU_counts[x] >= threshold else 'Other')
sns.countplot(x='BU_grouped', data=df, order=df['BU_grouped'].value_counts().index)
plt.title('Frequency of BU (Grouped)')
plt.xlabel('BU')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
Model_counts = df['Model'].value_counts()
threshold = 5
df['Model_grouped'] =df['Model'].apply(lambda x: x if Model_counts[x] >= threshold else 'Other')
sns.countplot(x='Model_grouped', data=df, order=df['Model_grouped'].value_counts().index)
plt.title('Frequency of Model (Grouped)')
plt.xlabel('Model')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Step1:
import numpy as np
import scipy.stats as stats
#step2:
sample_mean=80
population_mean=100
population_std=15
sample_size=50
#n=50
alpha=0.05
#step3:
z_score=(sample_mean-population_mean)/(population_std/np.sqrt(50))
print(z_score)
print(' The Z-Score is:',z_score)
# Z -TEST IS EVALUATED USING Z SCORE
#Step4
#critical Z score,ppf inverse of cdf,to find z critical we use ppf,percent point Function
z_critical=stats.norm.ppf(1-alpha)
#step5
if z_score > z_critical:
    print(" Reject Null HypotheSIS")
else:
    print("ACCEPT nULL HYPOTHESIS")
    #END THIS HYPOTHESIS TESTING
#Step6:
# PVALUE IS  CALCULATED USING CDF
p_value=1-stats.norm.cdf(z_score)
#step7:
if p_value < alpha:
    print(" Reject Null HypotheSIS")
else:
    print("ACCEPT nULL HYPOTHESIS")
```
